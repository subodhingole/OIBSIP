Importing necessary libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

data = pd.read_csv(
    "https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv"
)
print(data.head())
```

Checking if the dataset contains missing values or not.

```python
print(data.isnull().sum())
```

```python
data.columns = [
    "States",
    "Date",
    "Frequency",
    "Estimated Unemployment Rate",
    "Estimated Employed",
    "Estimated Labour Participation Rate",
    "Region",
    "longitude",
    "latitude",
]
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr())
plt.show()
```

Analyzing the unemployment rate.

```python
data.columns = [
    "States",
    "Date",
    "Frequency",
    "Estimated Unemployment Rate",
    "Estimated Employed",
    "Estimated Labour Participation Rate",
    "Region",
    "longitude",
    "latitude",
]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.show()
```

Unemployment rate according to different regions in India.

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
```

SunBurst Plotting

```python
unemploment = data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(
    unemploment,
    path=["Region", "States"],
    values="Estimated Unemployment Rate",
    width=700,
    height=700,
    color_continuous_scale="RdY1Gn",
    title="Unemployment Rate in India",
)
figure.show()
```
